```python
# %load influxdb_write.py
client = connection()
bucket="pre_maint"

write_api = client.write_api(write_options=SYNCHRONOUS)
   
for value in np.random.random(10):
  value = value.astype(int)
  for location in ['beijing','shanghai','tianjin']:
    point = (
      Point("measurement1")
      .tag("location", location)
      .field("temperature", value)
    )
    write_api.write(bucket=bucket, org="bosch", record=point)
    #time.sleep(1) # separate points by 1 second

print(f'write is done')
```

```python
# %load influxdb_read.py
import pandas as pd
client = connection()
query_api = client.query_api()
csv_result = query_api.query_csv('from(bucket:"pre_maint") |> range(start: -10m)')

df = pd.DataFrame(data=csv_result)
df.columns = df.iloc[3]
df = df.iloc[4:,:]

#print(df)



```
